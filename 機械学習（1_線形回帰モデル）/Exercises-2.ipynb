```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib.font_manager import FontProperties as fp
JP = fp(fname=r"C:\WINDOWS\Fonts\BIZ-UDGothicR.ttc", size=14)

# 分析データとしてボストン住宅セットを使用
# DATAFILE = "https://archive.ics.uci.edu/ml/machine-learning-databases/housing/housing.data"
DATAFILE = "housing.data"
COLNAMES = ["CRIM", "ZN", "INDUS", "CHAS", "NOX", "RM", "AGE",
            "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT", "MEDV"]
df = pd.read_table(DATAFILE, sep="\s+", names=COLNAMES)
df.head()
```

```python
import seaborn as sns

# 部屋数と住宅価格の相関
ax = sns.regplot(x='RM', y='MEDV', data=df)
ax.set_xlabel("部屋数", fontproperties=JP)
ax.set_ylabel("住宅価格", fontproperties=JP)
plt.show()
```

```python

# 犯罪発生率と住宅価格の相関
ax = sns.regplot(x='CRIM', y='MEDV', data=df)
ax.set_xlabel("犯罪発生率", fontproperties=JP)
ax.set_ylabel("住宅価格", fontproperties=JP)
plt.show()
```
